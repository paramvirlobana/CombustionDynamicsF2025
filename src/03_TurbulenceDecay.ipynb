```python
import os
import numpy as np
import pandas as pd
import cantera as ct
from time import time
import argparse
import matplotlib.pyplot as plt
```

```python
# Directories
DIR     = os.getcwd()
DATA    = os.path.join(DIR, "data", "csv")
```

```python
def reader(csv:str):
    data = np.loadtxt(csv, delimiter=',', unpack=True)
    return data
from scipy.interpolate import interp1d
def func_interpolate(xData, yData, kind='linear'):
    return interp1d(xData, yData, kind=kind)
```

```python
x2, y2 = reader(os.path.join(DATA, "fig5-35.csv"))
f2 = func_interpolate(x2, y2)
```

```python
eta_range = np.linspace(0.007, 0.29, 101)
sigma = np.zeros_like(eta_range)

for e, eta in enumerate(eta_range):
    sigma[e] = f2(eta)
```

```python
plt.plot(eta_range,sigma)
plt.show()
```

```python
sigma0 = 0.25
sigma2 = (sigma*sigma)**2

plt.plot(eta_range,sigma)
plt.plot(eta_range, sigma2)
plt.show()
```
